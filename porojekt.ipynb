Prije nego počmemo treba importat biblioteke koje će se koristit

```python
import sqlite3
import json
import csv
import matplotlib.pyplot as plt
```

```python
json_putanja = r"C:\Users\VLASIC\Programiranje za analizu\zamjena_projekta\datoteka.json"
csv_putanja = r"C:\Users\VLASIC\Programiranje za analizu\zamjena_projekta\datoteka.csv"

def ucitaj_json(json_putanja):
    with open(json_putanja, encoding='utf-8') as json_datoteka:
        return json.load(json_datoteka)

def ucitaj_csv(csv_putanja):
    with open(csv_putanja, 'r') as csv_datoteka:
        csv_citatelj= csv.reader(csv_datoteka)
        return [redak for redak in csv_citatelj]

def ispisi_prvih_n_redaka(json_putanja, n=5):
    podaci_json =ucitaj_json(json_putanja)
    print(f"Sadržaj prvih {n} redaka datoteke u JSON formatu:")
    for redovi in podaci_json[:n]:
        print(json.dumps(redovi, indent = 1))
    return podaci_json[:n]


def ispisi_prvih_n_redaka_csv(csv_putanja, n = 5):
    csv_podaci=ucitaj_csv(csv_putanja)
    print(f"\nSadržaj prvih {n} redaka datoteke u CSV formatu:")
    for i, redak in enumerate(csv_podaci):
        if i< n:
            print(redak)
    return csv_podaci[:n]

json_podaci_za_ispis=ispisi_prvih_n_redaka(json_putanja)
csv_podaci =ispisi_prvih_n_redaka_csv(csv_putanja)
json_podaci= ucitaj_json(json_putanja)
```

**Opis koda:**

**JSON Datoteka:**

- `json_putanja`: definira apsolutna putanja do JSON datoteke.
- `ispisi_prvih_n_redaka`: Funkcija koja prima putanju do JSON datoteke i broj redaka (n) koje želite ispisati tako da ne ispišemo svih 100 redaka radi preglednosti.
- `with open(json_putanja, encoding='utf-8') as json_datoteka`: Otvara JSON datoteku s kodiranjem UTF-8 kako bi podržala specifične znakove.
- `podaci_json = json.load(json_datoteka)`: Učitava podatke iz JSON datoteke.
- Nakon toga, koristi se petlja za ispis prvih n redaka podataka u format JSON.

**CSV Datoteka:**

- `csv_putanja`: definira apsolutna putanja do CSV datoteke.
- `ispisi_prvih_n_redaka_csv`: Funkcija koja prima putanju do CSV datoteke i broj redaka (n) koje želite ispisati tako da ne ispišemo svih 100 redaka radi preglednosti.
- `with open(csv_putanja, 'r') as csv_datoteka`: Otvara CSV datoteku za čitanje.
- `csv_citatelj = csv.reader(csv_datoteka)`: Stvara čitatelja za čitanje redaka iz CSV datoteke.
- Nakon toga, koristi se petlja za ispis prvih n redaka podataka u formatu CSV.

**Pozivi Funkcija:**

- Pozivaju se obje funkcije (`ispisi_prvih_n_redaka` i `ispisi_prvih_n_redaka_csv`) s definiranim putanjama do datoteka.
- Ovdje se pretpostavlja da će se ispisati prvih 5 redaka, ali to se može prilagoditi promjenom vrijednosti argumenta n prilikom poziva funkcija.




**Razlika:**


1. **Hijerarhija:**
   - JSON omogućava hijerarhijsku strukturu s ključevima i podobjektima.
   - CSV nema hijerarhiju; svi podaci su ravni i nalaze se u stupcima.

2. **Atributi:**
   - JSON ima više atributa kao što su "address" i "region".
   - CSV koristi samo atribute navedene kao zaglavlje stupca.

3. **Tipovi Podataka:**
   - JSON eksplicitno prikazuje tipove podataka, poput stringova i brojeva.
   - CSV pretpostavlja da su svi podaci stringovi, a tipovi podataka se moraju tumačiti iz konteksta.

4. **Čitljivost:**
   - JSON je često čitljiviji zbog svoje strukture i jasnoće.

### Preptocesiranje

Proces pretprocesiranja podataka ima ključnu ulogu u pripremi podataka za analizu ili modeliranje. U ovom kontekstu, napisani kod ispod pokriva nekoliko ključnih koraka pretprocesiranja:

**Čišćenje Podataka**

1. **Uklanjanje Nepotrebnih Značajki:**
   - Određene značajke (stupci) su izbačene iz podataka kako bi se smanjila dimenzionalnost i fokusiralo se samo na bitne informacije.

2. **Promjena Naziva Stupca:**
   - Naziv stupca "numberrange" je promijenjen u "Dob" kako bi se poboljšala čitljivost ili usklađenost s drugim datasetima.

**Transformacija Podataka**

3. **Provjeri i Rukuj Nedostajućim Vrijednostima:**
   - Iteracija kroz podatke kako bi se provjerile nedostajuće ili prazne vrijednosti, uz primjenu strategije rukovanja (u ovom slučaju zamjena nulom).
   - Ovaj korak osigurava da podaci budu ujednačeni i spremni za daljnju analizu.

```python
#prvi dio koda

csv_podaci=[]
with open(csv_putanja, newline='') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        del row['postalZip']
        del row['phone']
        del row['email']
        csv_podaci.append(row)


for i in json_podaci:
    if 'postalZip' in i:
        del i['postalZip']
    if 'phone' in i:
        del i['phone']
    if 'email' in i:
        del i['email']
    if 'region' in i:
        del i['region']
    if 'list' in i:
        del i['list']
    if 'address' in i:
        del i['address']

#drugi dio koda
for i in json_podaci:
    if 'numberrange' in i:
        i['Dob'] = i.pop('numberrange')

for i in csv_podaci:
    i['Dob'] = i.pop('numberrange')

print("Ažurirani JSON podaci:")
for i in json_podaci:
    print(i)

print("Ažurirani CSV podaci:")
for i in csv_podaci:
    print(i)
    
#treći dio koda
def provjeri_i_rukuj_nedostajucim(data):
    ima_nedostajucih=False
    for i in data:
        for kljuc,value in i.items():
            if value is None or value == '':
                print(f'Nedostajuća vrijednost u retku za ključ "{kljuc}"')
                i[kljuc] = 0
                ima_nedostajucih= True
    
    if not ima_nedostajucih:
        print(" Nema nedostajućih vrijednosti.")
        
print("# Provjeri i rukuj nedostajućim vrijednostima")
provjeri_i_rukuj_nedostajucim(json_podaci)
provjeri_i_rukuj_nedostajucim(csv_podaci)
```

U prvom dijelu koda, učitavanje CSV Podataka i brisanje nepotrebnih stupaca

- Ovdje se otvara CSV datoteka i čita koristeći csv.DictReader, što znači da svaki redak iz datoteke postaje rječnik. Zatim se brisu nepotrebni stupci poput 'postalZip', 'phone' i 'email', a zatim se dodaju preostali podaci u listu csv_podaci.


U drugom dijelu koda brisanje nepotrebnih stupaca iz JSON podataka

- Ovdje se prolazi kroz JSON podatke i brišu se određeni nepotrebni stupci poput 'postalZip', 'phone', 'email', 'region', 'list' i 'address'.

U trećem dijelu koda dodavanje novog stupca i provjera nedostajućih vrijednosti

- U ovom dijelu koda dodaju se novi stupci 'Dob' u JSON i CSV podatke. Također, provjerava se i rukuje nedostajućim vrijednostima ako takvih ima putem funkcije provjeri_i_rukuj_nedostajucim.


### Integracije podataka iz više skupova


Ovaj dio koda vrši spajanje podataka iz dvije različite strukture, json_podaci i csv_podaci, u novu kolekciju podataka nazvanu zajednicki_podaci. Nakon što su podaci spojeni, koristi se petlja za ispisivanje svakog elementa zajednicki_podaci.

- `for i in zajednicki_podaci: print(i)`: Ova petlja prolazi kroz svaki element u listi zajednicki_podaci i ispisuje ga. Svaki element može biti rječnik, redak iz CSV-a ili druga struktura podataka koja je prisutna u obje liste.

## Pohrane integriranog skupa podataka u bazu

Ovaj dio koda odnosi se na povezivanje s SQLite bazom podataka, stvaranje tablice integrirani_podaci i umetanje podataka iz liste zajednicki_podaci u tu tablicu.

```python
baza_podataka= sqlite3.connect('integrirani_podaci.db')
cursor =baza_podataka.cursor()
cursor.execute('''
    CREATE TABLE IF NOT EXISTS integrirani_podaci (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT,
        country TEXT,
        currency DECIMAL,
        Dob INTEGER
    )
''')

for i in zajednicki_podaci:
    cursor.execute('''
        INSERT INTO integrirani_podaci (name, country, currency, Dob)
        VALUES (?, ?, ?, ?)
    ''', (i['name'], i['country'], i['currency'], i['Dob']))


baza_podataka.commit()
baza_podataka.close()

print("Podaci su pohranjeni u bazu podataka.")
```

- `import sqlite3`: Ova linija uvozi modul sqlite3, koji omogućuje rad s bazama podataka SQLite u Pythonu.

- `baza_podataka = sqlite3.connect('integrirani_podaci.db')`: Ovdje se uspostavlja veza s bazom podataka. Ako datoteka 'integrirani_podaci.db' ne postoji, stvorit će se nova baza podataka. Ako veza s bazom podataka već postoji, koristi se postojeća veza.

- `cursor = baza_podataka.cursor()`: Stvara se objekt cursor, koji se koristi za izvršavanje SQL upita na bazi podataka.

- `cursor.execute('''CREATE TABLE IF NOT EXISTS integrirani_podaci ... ''')`: Izvršava se SQL upit koji stvara tablicu integrirani_podaci u bazi podataka ako već ne postoji. Tablica ima stupce id (kao primarni ključ), name, country, currency, i Dob.

- `for i in zajednicki_podaci: cursor.execute('''INSERT INTO integrirani_podaci ... ''', (i['name'], i['country'], i['currency'], i['Dob']))`: Koristi se petlja za ubacivanje podataka iz liste zajednicki_podaci u tablicu integrirani_podaci. Svaki element iz liste predstavlja jedan redak u tablici, a vrijednosti se ubacuju u odgovarajuće stupce.

- `baza_podataka.commit()`: Nakon što su svi podaci ubačeni, poziva se commit kako bi se promjene trajno spremljene u bazi podataka.

- `baza_podataka.close()`: Zatvara se veza s bazom podataka nakon što su svi podaci pohranjeni.

- `rint("Podaci su pohranjeni u bazu podataka.")`: Ispisuje se poruka koja potvrđuje da su podaci uspješno pohranjeni u bazu podataka.

Ovaj dio koda omogućuje trajno pohranjivanje integriranih podataka u SQLite bazu podataka kako bi se olakšala daljnja uporaba i analiza.

Za dodanu korisnost i jednostavnost dodan je kod za brisanje svega što se nalazi u bazi tako da možemo ponovno spremati.

```python
baza_podataka= sqlite3.connect('integrirani_podaci.db')
cursor= baza_podataka.cursor()

cursor.execute('DELETE FROM integrirani_podaci')

baza_podataka.commit()
baza_podataka.close()

print("Podaci su ispraznjeni iz tablice integrirani_podaci.")
```

Ovaj dio koda izvršava operaciju brisanja svih podataka iz tablice integrirani_podaci u SQLite bazi podataka. 

Uspostavljanje veze s bazom podataka:
- U ovom koraku, otvaramo vezu s bazom podataka integrirani_podaci.db. To nam omogućuje komunikaciju s bazom podataka.

Brisanje podataka iz tablice:
- Izvršavamo SQL upit koji glasi DELETE FROM integrirani_podaci. Ovaj upit briše sve retke (podatke) iz tablice integrirani_podaci, ostavljajući praznu tablicu.

Spremanje promjena:
- Nakon što obavimo brisanje podataka, potrebno je spremiti te promjene u bazi podataka pomoću commit naredbe. To osigurava trajno brisanje podataka.

Zatvaranje veze s bazom podataka:
- Zatvaramo vezu s bazom podataka kako bismo oslobodili resurse nakon završetka operacije brisanja.

Ispis poruke:
- Na kraju, ispisuje se poruka koja potvrđuje da su podaci uspješno izbrisani iz tablice integrirani_podaci.

```python
baza_podataka= sqlite3.connect('integrirani_podaci.db')

cursor =baza_podataka.cursor()
cursor.execute('SELECT * FROM integrirani_podaci')

rezultati = cursor.fetchall()

for redak in rezultati:
    print(redak)

baza_podataka.close()
```

Da bi pregledali šta se nalazi u bazu izvršili smo prije navedeni kod.

Izvršavanje SQL upita SELECT:
- Pomoću execute metode izvršavamo SQL upit SELECT * FROM integrirani_podaci. Ovaj upit dohvaća sve retke iz tablice integrirani_podaci.

Dohvaćanje rezultata:
- Nakon izvršavanja upita, koristimo fetchall metodu za dohvaćanje svih rezultata upita. Rezultati su pohranjeni u varijablu rezultati.

Ispis rezultata:
- Koristimo petlju for za prolazak kroz rezultate i ispisivanje svakog retka.

Zatvaranje veze s bazom podataka:
- Na kraju, zatvaramo vezu s bazom podataka kako bismo oslobodili resurse nakon završetka izvršavanja upita.

##  Razvoj REST API-ja pomoću Flask-a

Razvoj REST API-ja (Application Programming Interface) predstavlja ključnu komponentu suvremenih softverskih sustava, omogućujući interakciju između različitih aplikacija i platformi. REST (Representational State Transfer) predstavlja arhitekturalni stil koji se često koristi za izgradnju web usluga. Flask, kao lagani web framework za Python, pruža jednostavan i efikasan način za postavljanje REST API-ja.

Zašto razvijati REST API pomoću Flask-a?

- Lagano učenje i korištenje: Flask je jednostavan za učenje, što ga čini pogodnim i za početnike u web razvoju.
- Fleksibilnost i proširivost: Flask pruža minimalan set alata koji omogućuju razvoj web aplikacija prema vlastitim potrebama, pružajući istovremeno mogućnost proširivosti prema kompleksnijim projektima.
- Široka zajednica i dokumentacija: Flask ima aktivnu zajednicu korisnika i opsežnu dokumentaciju, što olakšava rješavanje problema i pronalaženje odgovora na pitanja.

Prvi koraci: Postavljanje Flask-a

Prije nego što krenemo u izradu REST API-ja, prvi korak je instalacija Flask-a. Ova instalacija omogućuje nam jednostavno postavljanje web servera i razvoj prvih ruta za naš API. Kroz naredbu pip install Flask, dobivamo pristup svim potrebnim alatima za pokretanje našeg Flask projekta.

Nakon uspješne instalacije možemo započeti.

```python
from flask import Flask, jsonify

app =Flask(__name__)

@app.route('/podaci', methods=['GET'])
def get_podaci():
    baza_podataka = sqlite3.connect('integrirani_podaci.db')
    cursor = baza_podataka.cursor()
    cursor.execute('SELECT * FROM integrirani_podaci')

    rezultati =cursor.fetchall()
    response =jsonify({'podaci': rezultati})
    
    baza_podataka.close()
    return response

if __name__ == '__main__':
    app.run(debug=True, use_reloader=False)
```

Za pristup link http://127.0.0.1:5000/podaci

Ovaj kod predstavlja osnovnu implementaciju Flask web aplikacije koja omogućuje pristup podacima putem HTTP GET zahtjeva na ruti /podaci. 

- `from flask import Flask, jsonify`: Ovdje se uvoze potrebni moduli. Flask se koristi za izradu web aplikacije, dok jsonify pomaže u pretvaranju rezultata upita u JSON format.

- `app = Flask(name)`: Stvara se instanca Flask aplikacije, nazvana app.

- `@app.route('/podaci', methods=['GET'])`: Ovaj dekorator postavlja rutu /podaci za HTTP GET zahtjev. To znači da će ova ruta biti aktivirana kada klijent pošalje GET zahtjev na tu adresu.

- `def get_podaci(): ... `: Ova funkcija definira što će se dogoditi kada stigne GET zahtjev na rutu /podaci. Unutar funkcije:

- Stvara se veza s SQLite bazom podataka imenovanom 'integrirani_podaci.db'.
- Stvara se cursor objekt za izvršavanje SQL upita.
- Izvršava se SQL upit koji dohvaća sve podatke iz tablice integrirani_podaci.
- Svi rezultati upita se dohvaćaju pomoću fetchall metode.
- Rezultati se pretvaraju u JSON format pomoću jsonify funkcije.
- Zatvara se veza s bazom podataka.
- JSON odgovor šalje se klijentu.

- `if name == 'main'`: Ova linija osigurava da se aplikacija pokreće samo ako je izravno pokrenuta, a ne ako je uvezena u drugi modul.

- `app.run(debug=True, use_reloader=False)`: Pokreće Flask aplikaciju. Postavke debug=True omogućuju automatsko osvježavanje aplikacije prilikom promjena u kodu, dok use_reloader=False sprječava dodatno pokretanje aplikacije prilikom izmjena koda.

Flask aplikacija omogućuje integraciju s različitim alatima i uslugama koje podržavaju HTTP protokol. Ova fleksibilnost omogućuje širok spektar mogućnosti, uključujući povezivanje s frontend sučeljem, mobilnim aplikacijama ili analitičkim alatima. Flask je lagan i prilagodljiv okvir koji olakšava proširenje i prilagodbu. Dodavanjem novih ruta ili ažuriranjem postojećih, možemo prilagoditi aplikaciju prema promjenama u potrebama korisnika ili dodavanju novih funkcionalnosti.

```python
dob_values =[redak[4] for redak in rezultati] 
prosjek_doba= sum(dob_values)/ len(dob_values)
print(f"Srednja vrijednost Dob: {prosjek_doba}")
```

Dohvaćanje Vrijednosti Doba iz Rezultata:

- `dob_values = [redak[4] for redak in rezultati]`: Pomoću ovog izraza dobivamo listu dobnih vrijednosti iz rezultata SQL upita. Indeks 4 odgovara stupcu u kojem se nalazi informacija o dobi.
Izračun Srednje Vrijednosti:

- `prosjek_doba = sum(dob_values) / len(dob_values)`: Računamo srednju vrijednost tako što zbrojimo sve dobnih vrijednosti i podijelimo taj zbroj s ukupnim brojem dobnih podataka. Ovaj rezultat predstavlja prosječnu dob u skupu.


Nakon stjecanja uvida u prosječnu dob kupaca pametnih telefona, ulazimo u fazu kompleksnije analize odnosa podataka. Ova faza omogućava nam istraživanje dubljih uzročno-posljedičnih veza, identificiranje trendova i bolje razumijevanje raznolikosti podataka. Kroz složenije prikaze, stvorit ćemo vizualizacije koje će pružiti dodatne insights-e i potaknuti dublju analizu karakteristika našeg skupa podataka. Ovaj korak koristan je za donošenje informiranih poslovnih odluka i oblikovanje strategija prilagođenih specifičnostima tržišta pametnih telefona.

```python
dob_values=[redak[4] for redak in rezultati]  

plt.hist(dob_values, bins= 20, color ='blue', edgecolor ='black')
plt.xlabel('Dob')
plt.ylabel('Broj unosa')
plt.title('Histogram dobi')
plt.show()
```

Objašnjenje koda:

- `dob_values`: Stvara listu vrijednosti dobi iz rezultata upita. Vrijednosti se uzimaju iz petog stupca (redak[4]) koji predstavlja dob u skupu podataka.

- `plt.hist(dob_values, bins=20, color='blue', edgecolor='black')`: Funkcija hist se koristi za stvaranje histograma. Parametar dob_values predstavlja podatke koje želimo analizirati, bins=20 određuje broj stupnjeva (pravokutnika) u histogramu, a color i edgecolor određuju boju ispune i rubova histograma.

- `plt.xlabel('Dob') i plt.ylabel('Broj unosa')`: Postavljaju oznake na x i y osi, gdje je x oznaka "Dob", a y oznaka "Broj unosa".

- `plt.title('Histogram dobi')`: Postavlja naslov histograma.

- `plt.show()`: Prikazuje stvoreni histogram.
    

Analiziranjem ovog histograma možemo dobivenim vizualnim prikazom uvidjeti raspodjelu dobnih skupina unutar našeg skupa podataka. Histogram nam omogućuje vizualno identificiranje frekvencije određenih dobnih grupa te nam pruža uvid u to gdje se najviše koncentriraju podaci.

U konkretnom primjeru, zapažamo da je vrhunac histograma smješten oko dobi od 45 godina. Ova dominacija u dobi sugerira da je dob od 45 godina najčešća među korisnicima koji su kupili pametne uređaje. To nam pruža vrijedan uvid u ciljnu dobnu skupinu koja najviše preferira ove uređaje, što može biti korisno pri planiranju marketinških strategija ili prilagodbi proizvoda prema preferencijama ove specifične dobne skupine.

- Iteracija kroz rezultate upita: Kroz svaki redak rezultata upita iz baze podataka prolazi se pomoću petlje.

- Dohvaćanje podataka o zemlji: Za svaki redak dohvaća se vrijednost stupca koji predstavlja zemlju.

- Priprema podataka za kružni dijagram: Stvaraju se liste `sizes` i `labels` kako bi se omogućila vizualizacija udjela zemalja.

- Postavljanje veličine slike kružnog dijagrama: Postavlja se veličina slike na 15x11 inča kako bi se omogućila bolja vidljivost.
